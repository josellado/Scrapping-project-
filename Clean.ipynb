## HARRY POTTER DATA FRAME 


```python
import pandas as pd
import numpy as np
import re
from dotenv import load_dotenv
load_dotenv()
import requests
from bs4 import BeautifulSoup
import os
```

```python
harry_potter = pd.read_csv("input/harry_potter/Characters.csv", sep =';', index_col=0, encoding ='latin')
harry_potter
```

> Filling the empty information with Unknown


```python
harry_potter= harry_potter.fillna('Unknown')
harry_potter.head()
```

> Prociding to Delete Columns that are not neccesary to my analysis

```python
harry_potter = harry_potter.drop(columns = ['Birth', 'Death', "Loyalty"])
harry_potter.shape
```

```python
harry_potter.isnull().sum()
```

 > Sort Values by Alphabetical Order based on 'Name' column

```python
harry_potter.sort_values('Name')
```

> Enriching  the data set from potterapi. To do so, I divided up the URL in different parts



```python
base_url = "https://www.potterapi.com/v1/"

end_url = "sortinghat"


url = base_url + end_url
response = requests.get(url)
data = response.json()
data
```

> Importing my ApiKey from an .env file to protect personal information




```python
apiKey = os.getenv("APIKEY")
```

```python
end_url = 'characters'
url = base_url + end_url

url = url + "?key=" + apiKey
url

response = requests.get(url)
data = response.json()
data[0]
```

> Creating a new data Frame with the information of the API

```python
df = pd.DataFrame(data)
df
```

```python
df.shape
```

> Im only gonna stay with the columns that enrich my Harry Potter Data set

```python
df.columns
```

```python
army = df.drop(columns = ['_id', 'role', 'house', 'school', '__v','deathEater', 'bloodStatus','species', 'boggart', 'alias', 'animagus', 'wand', 'patronus'])
      
```

```python
army
```

```python
army.sort_values('name')
```

> Merging the two data frames together by Name.

```python
final_data =harry_potter.merge(army,how='left', left_on='Name', right_on='name')
final_data.sort_values('Name')
```

```python
final_data.isnull().sum()
```

> Changing all the Nan Values to Delete, in the following lines I will Delete all rows containig these word

```python
final_data= final_data.fillna('delete')
```

```python
final_data.head()
```

```python
searchfor = ['delete']
army = final_data[~final_data.name.str.contains('|'.join(searchfor))]
army.sort_values("Name")
```

> Getting rid off the duplicate columm name

```python
army = army.drop(columns = ['name'])
army = army.sort_values("Name")
```
